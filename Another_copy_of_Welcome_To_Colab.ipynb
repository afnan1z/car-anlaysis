```
!pip install -q ultralytics kagglehub opencv-python

import os
import cv2
import numpy as np
from ultralytics import YOLO
import kagglehub

# تحميل الداتا من Kaggle
dataset_path = kagglehub.dataset_download("aryashah2k/highway-traffic-videos-dataset")
videos_dir = os.path.join(dataset_path, "video")

all_videos = sorted(os.listdir(videos_dir))

model = YOLO("yolov8n.pt")

def score_video(video_path):
    cap = cv2.VideoCapture(video_path)
    if not cap.isOpened():
        return 0

    car_counts = []
    motion_scores = []
    prev_centers = None

    for _ in range(30):
        ret, frame = cap.read()
        if not ret:
            break

        results = model(frame, verbose=False)[0]
        centers = []

        for box in results.boxes:
            cls = int(box.cls)
            # سيارات / شاحنات / باصات / موتوسيكلات
            if cls in [2, 3, 5, 7]:
                x1, y1, x2, y2 = box.xyxy[0]
                cx = float((x1 + x2) / 2)
                cy = float((y1 + y2) / 2)
                centers.append((cx, cy))

        car_counts.append(len(centers))

        if prev_centers is not None and len(centers) == len(prev_centers):
            motion = np.mean([
                np.linalg.norm(np.array(centers[i]) - np.array(prev_centers[i]))
                for i in range(len(centers))
            ])
            motion_scores.append(motion)

        prev_centers = centers

    cap.release()

    if len(car_counts) == 0:
        return 0

    avg_cars = np.mean(car_counts)
    avg_motion = np.mean(motion_scores) if len(motion_scores) > 0 else 0

    return avg_cars * 1.5 + avg_motion * 0.5


best_video = None
best_score = -1

for i, v in enumerate(all_videos[:80]):
    video_path = f"{videos_dir}/{v}"
    score = score_video(video_path)
    print(f"{i}: {v}   → Score: {score:.2f}")

    if score > best_score:
        best_score = score
        best_video = v

selected_video_path = f"{videos_dir}/{best_video}"

print("Best video path:")
print(selected_video_path)
```

```
!pip install -q ultralytics kagglehub opencv-python scikit-learn pandas

import os
import math
import cv2
import numpy as np
from collections import deque
from ultralytics import YOLO
import kagglehub
from IPython.display import Video, display
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

dataset_path = kagglehub.dataset_download("aryashah2k/highway-traffic-videos-dataset")
video_dir = os.path.join(dataset_path, "video")

VIDEO_NAME = "cctv052x2004080518x01668.avi"
VIDEO_PATH = os.path.join(video_dir, VIDEO_NAME)

print("Dataset:", dataset_path)
print("Video dir:", video_dir)
print("Video:", VIDEO_PATH)

cap = cv2.VideoCapture(VIDEO_PATH)
if not cap.isOpened():
    raise RuntimeError("Cannot open video, check VIDEO_NAME.")

fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out_path = "predicted_highway_risks.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(out_path, fourcc, fps, (w, h))

print(f"Size: {w}x{h} | FPS: {fps:.2f}")

model = YOLO("yolov8n.pt")

RESTRICT_X_MIN, RESTRICT_X_MAX = 0.42, 0.52
RESTRICT_Y_MIN, RESTRICT_Y_MAX = 0.00, 1.00

HIGH_DENSITY_THRESHOLD = 25
IDLE_FRAMES_THRESHOLD = 25
SUDDEN_STOP_SPEED_DROP = 8.0
MAX_MATCH_DIST = 50
PREDICTION_WINDOW = 30

class RiskPredictor:
    def __init__(self):
        self.scaler = StandardScaler()
        self.classifier = RandomForestClassifier(n_estimators=100, random_state=42)
        self.is_trained = False
        self.feature_history = []
        self.label_history = []

    def extract_features(self, tracks, current_frame_features):
        features = []
        if len(tracks) > 0:
            speeds = [t.prev_speed for t in tracks.values() if len(t.positions) > 1]
            features.append(np.mean(speeds) if speeds else 0)
            features.append(np.std(speeds) if speeds else 0)
            features.append(len(tracks))

            stationary_ratio = sum(1 for t in tracks.values() if t.stationary_frames > 5) / len(tracks)
            features.append(stationary_ratio)

            if len(tracks) > 1:
                speed_changes = []
                for t in tracks.values():
                    if len(t.positions) > 3:
                        recent_speeds = []
                        for i in range(-3, -1):
                            if i < -len(t.positions):
                                continue
                            x1, y1 = t.positions[i]
                            x2, y2 = t.positions[i + 1] if i + 1 < 0 else t.positions[-1]
                            recent_speeds.append(math.hypot(x2 - x1, y2 - y1))
                        if recent_speeds:
                            speed_changes.append(np.std(recent_speeds))
                features.append(np.mean(speed_changes) if speed_changes else 0)
            else:
                features.append(0)
        else:
            features.extend([0, 0, 0, 0, 0])

        features.extend(current_frame_features)
        return np.array(features)

    def train_predictor(self):
        if len(self.feature_history) > 50:
            X = np.array(self.feature_history[:-PREDICTION_WINDOW])
            y = np.array(self.label_history[PREDICTION_WINDOW:])
            if len(np.unique(y)) > 1:
                X_scaled = self.scaler.fit_transform(X)
                self.classifier.fit(X_scaled, y)
                self.is_trained = True

    def predict_risk(self, features):
        if len(self.feature_history) < 10:
            return 0.1, "collecting"

        recent_features = self.feature_history[-10:]
        trends = []
        for i in range(len(recent_features[0])):
            values = [f[i] for f in recent_features]
            slope = (values[-1] - values[0]) / len(values)
            trends.append(slope)

        risk_score = 0.0

        if features[0] < 2.0 and trends[0] < -0.1:
            risk_score += 0.3
        if features[1] > 3.0:
            risk_score += 0.2
        if features[2] > HIGH_DENSITY_THRESHOLD * 0.8:
            risk_score += 0.2
        if features[3] > 0.3:
            risk_score += 0.3

        if self.is_trained:
            features_scaled = self.scaler.transform([features])
            proba = self.classifier.predict_proba(features_scaled)[0]
            if len(proba) > 1:
                risk_score += proba[1] * 0.5

        risk_score = max(0.0, min(1.0, risk_score))

        if risk_score > 0.7:
            risk_level = "CRITICAL"
        elif risk_score > 0.5:
            risk_level = "HIGH"
        elif risk_score > 0.3:
            risk_level = "MEDIUM"
        else:
            risk_level = "LOW"

        return risk_score, risk_level

class AdvancedTrack:
    def __init__(self, track_id, cx, cy):
        self.id = track_id
        self.positions = deque(maxlen=20)
        self.positions.append((cx, cy))
        self.predicted_positions = []
        self.speeds = deque(maxlen=10)
        self.accelerations = deque(maxlen=10)
        self.stationary_frames = 0
        self.prev_speed = 0.0
        self.sudden_stop_flagged = False
        self.long_idle_flagged = False
        self.restricted_flagged = False
        self.last_bbox = None
        self.predicted_collision = False

    def update(self, cx, cy):
        self.positions.append((cx, cy))
        if len(self.positions) >= 2:
            (x1, y1), (x2, y2) = list(self.positions)[-2], list(self.positions)[-1]
            dist = math.hypot(x2 - x1, y2 - y1)

            if len(self.speeds) > 0:
                acceleration = dist - self.speeds[-1]
                self.accelerations.append(acceleration)

            self.speeds.append(dist)
            self.prev_speed = dist
            self.predict_future_path()

            if dist < 3.0:
                self.stationary_frames += 1
            else:
                self.stationary_frames = 0

    def predict_future_path(self):
        if len(self.positions) < 5:
            self.predicted_positions = []
            return

        positions = list(self.positions)
        x_coords = [p[0] for p in positions[-5:]]
        y_coords = [p[1] for p in positions[-5:]]

        if len(x_coords) >= 2:
            dx = np.mean([x_coords[i + 1] - x_coords[i] for i in range(len(x_coords) - 1)])
            dy = np.mean([y_coords[i + 1] - y_coords[i] for i in range(len(y_coords) - 1)])

            last_pos = positions[-1]
            self.predicted_positions = []
            for i in range(1, 6):
                pred_x = last_pos[0] + dx * i
                pred_y = last_pos[1] + dy * i
                self.predicted_positions.append((pred_x, pred_y))

    def predict_collision_risk(self, other_tracks):
        if not self.predicted_positions or len(self.predicted_positions) < 3:
            return False, 0

        for tid, other_track in other_tracks.items():
            if tid == self.id or not other_track.predicted_positions:
                continue

            min_distance = float("inf")
            for i in range(min(3, len(self.predicted_positions), len(other_track.predicted_positions))):
                dist = math.hypot(
                    self.predicted_positions[i][0] - other_track.predicted_positions[i][0],
                    self.predicted_positions[i][1] - other_track.predicted_positions[i][1],
                )
                min_distance = min(min_distance, dist)

            if min_distance < 20:
                self.predicted_collision = True
                return True, 1.0 - (min_distance / 20)

        return False, 0

    def last_position(self):
        return self.positions[-1] if self.positions else (0, 0)

def detect_lane_changes(track, frame_index):
    if len(track.positions) < 10:
        return False

    positions = list(track.positions)
    recent_positions = positions[-5:]
    older_positions = positions[-10:-5]

    if len(recent_positions) < 2 or len(older_positions) < 2:
        return False

    recent_dx = np.mean([recent_positions[i + 1][0] - recent_positions[i][0] for i in range(len(recent_positions) - 1)])
    recent_dy = np.mean([recent_positions[i + 1][1] - recent_positions[i][1] for i in range(len(recent_positions) - 1)])
    older_dx = np.mean([older_positions[i + 1][0] - older_positions[i][0] for i in range(len(older_positions) - 1)])
    older_dy = np.mean([older_positions[i + 1][1] - older_positions[i][1] for i in range(len(older_positions) - 1)])

    angle_change = abs(math.atan2(recent_dy, recent_dx) - math.atan2(older_dy, older_dx))
    return angle_change > math.pi / 4

tracks = {}
next_track_id = 0
risk_predictor = RiskPredictor()

high_density_events = 0
sudden_stop_events = 0
long_idle_events = 0
restricted_zone_events = 0
predicted_collisions = 0
lane_change_warnings = 0

frame_index = 0
max_vehicles_in_frame = 0

global_risk_score = 0.0
global_risk_label = "collecting"

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_index += 1

    display_frame = frame.copy()
    results = model(frame, verbose=False)
    detections = []

    for r in results:
        for b in r.boxes:
            cls_id = int(b.cls[0])
            cls_name = model.names[cls_id]
            if cls_name not in ["car", "truck", "bus", "motorbike"]:
                continue

            x1, y1, x2, y2 = b.xyxy[0].cpu().numpy()
            cx = (x1 + x2) / 2
            cy = (y1 + y2) / 2
            detections.append((cx, cy, int(x1), int(y1), int(x2), int(y2), cls_name))

    unmatched_tracks = set(tracks.keys())
    assigned = []

    for det in detections:
        cx, cy, x1, y1, x2, y2, cls_name = det

        best_id = None
        best_dist = float("inf")

        for tid in unmatched_tracks:
            tx, ty = tracks[tid].last_position()
            dist = math.hypot(cx - tx, cy - ty)
            if dist < best_dist and dist < MAX_MATCH_DIST:
                best_dist = dist
                best_id = tid

        if best_id is not None:
            tr = tracks[best_id]
            tr.update(cx, cy)
            tr.last_bbox = (x1, y1, x2, y2)
            unmatched_tracks.remove(best_id)
            assigned.append(det + (best_id,))
        else:
            tracks[next_track_id] = AdvancedTrack(next_track_id, cx, cy)
            tracks[next_track_id].last_bbox = (x1, y1, x2, y2)
            assigned.append(det + (next_track_id,))
            next_track_id += 1

    for tid in list(unmatched_tracks):
        if tid in tracks:
            del tracks[tid]

    car_count = len(assigned)
    max_vehicles_in_frame = max(max_vehicles_in_frame, car_count)

    current_features = [
        car_count,
        np.mean([t.prev_speed for t in tracks.values()]) if tracks else 0,
        sum(1 for t in tracks.values() if t.stationary_frames > 5),
    ]

    frame_features = risk_predictor.extract_features(tracks, current_features)
    risk_predictor.feature_history.append(frame_features)
    risk_predictor.label_history.append(1 if car_count > HIGH_DENSITY_THRESHOLD else 0)

    if frame_index % 100 == 0 and frame_index > 100:
        risk_predictor.train_predictor()

    global_risk_score, global_risk_label = risk_predictor.predict_risk(frame_features)

    rx1 = int(RESTRICT_X_MIN * w)
    rx2 = int(RESTRICT_X_MAX * w)
    ry1 = int(RESTRICT_Y_MIN * h)
    ry2 = int(RESTRICT_Y_MAX * h)

    overlay = display_frame.copy()
    cv2.rectangle(overlay, (rx1, ry1), (rx2, ry2), (0, 0, 255), -1)
    display_frame = cv2.addWeighted(overlay, 0.15, display_frame, 0.85, 0)
    cv2.rectangle(display_frame, (rx1, ry1), (rx2, ry2), (0, 0, 255), 2)

    for det in assigned:
        cx, cy, x1, y1, x2, y2, cls_name, tid = det
        tr = tracks[tid]

        collision_risk, collision_prob = tr.predict_collision_risk(tracks)
        lane_change_risk = detect_lane_changes(tr, frame_index)

        color = (0, 255, 0)
        vehicle_risk_score = 0.1

        if collision_risk and collision_prob > 0.5:
            color = (0, 0, 255)
            vehicle_risk_score = 0.9
            if not hasattr(tr, "collision_warned"):
                predicted_collisions += 1
                tr.collision_warned = True

        elif lane_change_risk:
            color = (0, 255, 255)
            vehicle_risk_score = 0.6
            if not hasattr(tr, "lane_change_warned"):
                lane_change_warnings += 1
                tr.lane_change_warned = True

        elif tr.stationary_frames >= IDLE_FRAMES_THRESHOLD:
            color = (255, 0, 0)
            vehicle_risk_score = 0.65
            if not tr.long_idle_flagged:
                long_idle_events += 1
                tr.long_idle_flagged = True

        elif tr.prev_speed < 2.0 and len(tr.speeds) > 1 and max(tr.speeds) > SUDDEN_STOP_SPEED_DROP:
            color = (0, 165, 255)
            vehicle_risk_score = 0.7
            if not tr.sudden_stop_flagged:
                sudden_stop_events += 1
                tr.sudden_stop_flagged = True

        if tr.last_bbox is not None:
            bx1, by1, bx2, by2 = tr.last_bbox
            overlap = not (bx2 < rx1 or bx1 > rx2 or by2 < ry1 or by1 > ry2)
            if overlap:
                color = (255, 0, 255)
                vehicle_risk_score = 0.8
                if not tr.restricted_flagged:
                    restricted_zone_events += 1
                    tr.restricted_flagged = True

        cv2.rectangle(display_frame, (x1, y1), (x2, y2), color, 2)

        if tr.predicted_positions:
            for i, (px, py) in enumerate(tr.predicted_positions):
                radius = max(1, 5 - i)
                cv2.circle(display_frame, (int(px), int(py)), radius, (255, 255, 0), -1)

        risk_percent_text = f"{int(vehicle_risk_score * 100)}%"
        label_y = max(y1 - 5, 20)
        cv2.putText(display_frame, risk_percent_text, (x1, label_y),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 1)

    if car_count >= HIGH_DENSITY_THRESHOLD:
        high_density_events += 1

    hud_x1, hud_y1 = 10, 10
    hud_x2, hud_y2 = 300, 90

    roi = display_frame[hud_y1:hud_y2, hud_x1:hud_x2].copy()
    cv2.rectangle(display_frame, (hud_x1, hud_y1), (hud_x2, hud_y2), (0, 0, 0), -1)
    display_frame[hud_y1:hud_y2, hud_x1:hud_x2] = cv2.addWeighted(
        display_frame[hud_y1:hud_y2, hud_x1:hud_x2], 0.6, roi, 0.4, 0
    )

    risk_color = (0, 255, 0)
    if global_risk_score > 0.7:
        risk_color = (0, 0, 255)
    elif global_risk_score > 0.5:
        risk_color = (0, 165, 255)
    elif global_risk_score > 0.3:
        risk_color = (0, 255, 255)

    risk_percent_global = int(global_risk_score * 100)

    y = hud_y1 + 20
    cv2.putText(display_frame, f"Risk score: {risk_percent_global}%",
                (hud_x1 + 8, y), cv2.FONT_HERSHEY_SIMPLEX, 0.45, risk_color, 1)
    y += 18
    cv2.putText(display_frame, f"Risk level: {global_risk_label}",
                (hud_x1 + 8, y), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (255, 255, 255), 1)
    y += 18
    cv2.putText(display_frame, f"Vehicles: {car_count}",
                (hud_x1 + 8, y), cv2.FONT_HERSHEY_SIMPLEX, 0.45, (255, 255, 255), 1)

    out.write(display_frame)

cap.release()
out.release()

print("\n" + "=" * 55)
print("REPORT")
print("=" * 55)
print(f"Frames: {frame_index}")
print(f"Max Vehicles: {max_vehicles_in_frame}")
print(f"High Density Alerts: {high_density_events}")
print(f"Long Idle Alerts: {long_idle_events}")
print(f"Sudden Stop Alerts: {sudden_stop_events}")
print(f"Restricted Zone Alerts: {restricted_zone_events}")
print(f"Collision Warnings: {predicted_collisions}")
print(f"Lane Change Warnings: {lane_change_warnings}")
print("=" * 55)
print(f"Saved: {out_path}")

display(Video(out_path, embed=True, width=800))
```

```
!pip install -q ultralytics opencv-python scikit-learn pandas

import math
import cv2
import numpy as np
from collections import deque
from ultralytics import YOLO
from IPython.display import Video, display
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from google.colab import files
import warnings
warnings.filterwarnings('ignore')

uploaded = files.upload()
VIDEO_PATH = list(uploaded.keys())[0]
print("Video:", VIDEO_PATH)

cap = cv2.VideoCapture(VIDEO_PATH)
if not cap.isOpened():
    raise RuntimeError("Cannot open video")

fps = cap.get(cv2.CAP_PROP_FPS) or 30.0
w = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
h = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))

out_path = "predicted_highway_risks.mp4"
fourcc = cv2.VideoWriter_fourcc(*"mp4v")
out = cv2.VideoWriter(out_path, fourcc, fps, (w, h))
print(f"Size: {w}x{h} | FPS: {fps:.2f}")

model = YOLO("yolov8n.pt")

HIGH_DENSITY_THRESHOLD = 25
IDLE_FRAMES_THRESHOLD = 25
SUDDEN_STOP_SPEED_DROP = 8.0
MAX_MATCH_DIST = 50
PREDICTION_WINDOW = 30
WARMUP_FRAMES = 15
MIN_HISTORY_FOR_RISK = 6

class RiskPredictor:
    def __init__(self):
        self.scaler = StandardScaler()
        self.classifier = RandomForestClassifier(n_estimators=100, random_state=42)
        self.is_trained = False
        self.feature_history = []
        self.label_history = []

    def extract_features(self, tracks, current_frame_features):
        features = []
        if len(tracks) > 0:
            speeds = [t.prev_speed for t in tracks.values() if len(t.positions) > 1]
            features.append(np.mean(speeds) if speeds else 0)
            features.append(np.std(speeds) if speeds else 0)
            features.append(len(tracks))

            stationary_ratio = sum(1 for t in tracks.values() if t.stationary_frames > 5) / len(tracks)
            features.append(stationary_ratio)

            if len(tracks) > 1:
                speed_changes = []
                for t in tracks.values():
                    if len(t.positions) > 3:
                        recent_speeds = []
                        for i in range(-3, -1):
                            if i < -len(t.positions):
                                continue
                            x1, y1 = t.positions[i]
                            if i + 1 < 0:
                                x2, y2 = t.positions[i + 1]
                            else:
                                x2, y2 = t.positions[-1]
                            recent_speeds.append(math.hypot(x2 - x1, y2 - y1))
                        if recent_speeds:
                            speed_changes.append(np.std(recent_speeds))
                features.append(np.mean(speed_changes) if speed_changes else 0)
            else:
                features.append(0)
        else:
            features.extend([0, 0, 0, 0, 0])

        features.extend(current_frame_features)
        return np.array(features)

    def train_predictor(self):
        if len(self.feature_history) > 50:
            X = np.array(self.feature_history[:-PREDICTION_WINDOW])
            y = np.array(self.label_history[PREDICTION_WINDOW:])
            if len(np.unique(y)) > 1:
                X_scaled = self.scaler.fit_transform(X)
                self.classifier.fit(X_scaled, y)
                self.is_trained = True

    def predict_risk(self, features):
        if len(self.feature_history) < 10:
            return 0.1, "collecting"

        recent_features = self.feature_history[-10:]
        trends = []
        for i in range(len(recent_features[0])):
            values = [f[i] for f in recent_features]
            slope = (values[-1] - values[0]) / len(values)
            trends.append(slope)

        risk_score = 0.0
        if features[0] < 2.0 and trends[0] < -0.1:
            risk_score += 0.3
        if features[1] > 3.0:
            risk_score += 0.2
        if features[2] > HIGH_DENSITY_THRESHOLD * 0.8:
            risk_score += 0.2
        if features[3] > 0.3:
            risk_score += 0.3

        if self.is_trained:
            features_scaled = self.scaler.transform([features])
            proba = self.classifier.predict_proba(features_scaled)[0]
            if len(proba) > 1:
                risk_score += proba[1] * 0.5

        risk_score = max(0.0, min(1.0, risk_score))

        if risk_score > 0.7:
            risk_level = "CRITICAL"
        elif risk_score > 0.5:
            risk_level = "HIGH"
        elif risk_score > 0.3:
            risk_level = "MEDIUM"
        else:
            risk_level = "LOW"

        return risk_score, risk_level

class AdvancedTrack:
    def __init__(self, track_id, cx, cy):
        self.id = track_id
        self.positions = deque(maxlen=20)
        self.positions.append((cx, cy))
        self.predicted_positions = []
        self.speeds = deque(maxlen=10)
        self.accelerations = deque(maxlen=10)
        self.stationary_frames = 0
        self.prev_speed = 0.0
        self.sudden_stop_flagged = False
        self.long_idle_flagged = False
        self.last_bbox = None
        self.predicted_collision = False
        self.max_risk_score = 0.0

    def update(self, cx, cy):
        self.positions.append((cx, cy))
        if len(self.positions) >= 2:
            (x1, y1), (x2, y2) = list(self.positions)[-2], list(self.positions)[-1]
            dist = math.hypot(x2 - x1, y2 - y1)

            if len(self.speeds) > 0:
                acceleration = dist - self.speeds[-1]
                self.accelerations.append(acceleration)

            self.speeds.append(dist)
            self.prev_speed = dist
            self.predict_future_path()

            if dist < 3.0:
                self.stationary_frames += 1
            else:
                self.stationary_frames = 0

    def predict_future_path(self):
        if len(self.positions) < 5:
            self.predicted_positions = []
            return

        positions = list(self.positions)
        x_coords = [p[0] for p in positions[-5:]]
        y_coords = [p[1] for p in positions[-5:]]

        if len(x_coords) >= 2:
            dx = np.mean([x_coords[i + 1] - x_coords[i] for i in range(len(x_coords) - 1)])
            dy = np.mean([y_coords[i + 1] - y_coords[i] for i in range(len(y_coords) - 1)])

            last_pos = positions[-1]
            self.predicted_positions = []
            for i in range(1, 6):
                pred_x = last_pos[0] + dx * i
                pred_y = last_pos[1] + dy * i
                self.predicted_positions.append((pred_x, pred_y))

    def predict_collision_risk(self, other_tracks):
        if not self.predicted_positions or len(self.predicted_positions) < 3:
            return False, 0

        for tid, other_track in other_tracks.items():
            if tid == self.id or not other_track.predicted_positions:
                continue

            min_distance = float("inf")
            for i in range(min(3, len(self.predicted_positions), len(other_track.predicted_positions))):
                dist = math.hypot(
                    self.predicted_positions[i][0] - other_track.predicted_positions[i][0],
                    self.predicted_positions[i][1] - other_track.predicted_positions[i][1],
                )
                min_distance = min(min_distance, dist)

            if min_distance < 20:
                self.predicted_collision = True
                return True, 1.0 - (min_distance / 20)

        return False, 0

    def last_position(self):
        return self.positions[-1] if self.positions else (0, 0)

def detect_lane_changes(track):
    if len(track.positions) < 10:
        return False

    positions = list(track.positions)
    recent_positions = positions[-5:]
    older_positions = positions[-10:-5]

    if len(recent_positions) < 2 or len(older_positions) < 2:
        return False

    recent_dx = np.mean([recent_positions[i + 1][0] - recent_positions[i][0] for i in range(len(recent_positions) - 1)])
    recent_dy = np.mean([recent_positions[i + 1][1] - recent_positions[i][1] for i in range(len(recent_positions) - 1)])
    older_dx = np.mean([older_positions[i + 1][0] - older_positions[i][0] for i in range(len(older_positions) - 1)])
    older_dy = np.mean([older_positions[i + 1][1] - older_positions[i][1] for i in range(len(older_positions) - 1)])

    angle_change = abs(math.atan2(recent_dy, recent_dx) - math.atan2(older_dy, older_dx))
    return angle_change > math.pi / 4

tracks = {}
next_track_id = 0
risk_predictor = RiskPredictor()

high_density_events = 0
sudden_stop_events = 0
long_idle_events = 0
predicted_collisions = 0
lane_change_warnings = 0
frame_index = 0
max_vehicles_in_frame = 0

while True:
    ret, frame = cap.read()
    if not ret:
        break
    frame_index += 1
    display_frame = frame.copy()

    results = model(frame, verbose=False)
    detections = []

    for r in results:
        for b in r.boxes:
            cls_id = int(b.cls[0])
            cls_name = model.names[cls_id]
            if cls_name not in ["car", "truck", "bus", "motorbike"]:
                continue

            x1, y1, x2, y2 = b.xyxy[0].cpu().numpy()
            box_height = y2 - y1
            if y1 > h * 0.6 and box_height > h * 0.25:
                continue

            cx = (x1 + x2) / 2
            cy = (y1 + y2) / 2
            detections.append((cx, cy, int(x1), int(y1), int(x2), int(y2), cls_name))

    unmatched_tracks = set(tracks.keys())
    assigned = []

    for det in detections:
        cx, cy, x1, y1, x2, y2, cls_name = det
        best_id = None
        best_dist = float("inf")

        for tid in unmatched_tracks:
            tx, ty = tracks[tid].last_position()
            dist = math.hypot(cx - tx, cy - ty)
            if dist < best_dist and dist < MAX_MATCH_DIST:
                best_dist = dist
                best_id = tid

        if best_id is not None:
            tr = tracks[best_id]
            tr.update(cx, cy)
            tr.last_bbox = (x1, y1, x2, y2)
            unmatched_tracks.remove(best_id)
            assigned.append(det + (best_id,))
        else:
            tracks[next_track_id] = AdvancedTrack(next_track_id, cx, cy)
            tracks[next_track_id].last_bbox = (x1, y1, x2, y2)
            assigned.append(det + (next_track_id,))
            next_track_id += 1

    for tid in list(unmatched_tracks):
        if tid in tracks:
            del tracks[tid]

    car_count = len(assigned)
    max_vehicles_in_frame = max(max_vehicles_in_frame, car_count)

    current_features = [
        car_count,
        np.mean([t.prev_speed for t in tracks.values()]) if tracks else 0,
        sum(1 for t in tracks.values() if t.stationary_frames > 5),
    ]

    frame_features = risk_predictor.extract_features(tracks, current_features)
    risk_predictor.feature_history.append(frame_features)
    risk_predictor.label_history.append(1 if car_count > HIGH_DENSITY_THRESHOLD else 0)

    if frame_index % 100 == 0 and frame_index > 100:
        risk_predictor.train_predictor()

    global_risk_score, global_risk_label = risk_predictor.predict_risk(frame_features)

    for det in assigned:
        cx, cy, x1, y1, x2, y2, cls_name, tid = det
        tr = tracks[tid]

        if frame_index <= WARMUP_FRAMES or len(tr.positions) < MIN_HISTORY_FOR_RISK:
            vehicle_risk_score = 0.0
        else:
            collision_risk, collision_prob = tr.predict_collision_risk(tracks)
            lane_change_risk = detect_lane_changes(tr)

            positions = list(tr.positions)
            lateral_motion = 0.0
            if len(positions) >= 3:
                lateral_motion = positions[-1][0] - positions[-3][0]

            distance_to_wall = x1

            vehicle_risk_score = 0.0

            if collision_risk and collision_prob > 0.5:
                vehicle_risk_score = max(vehicle_risk_score, 0.9)
                if not hasattr(tr, "collision_warned"):
                    predicted_collisions += 1
                    tr.collision_warned = True

            if abs(lateral_motion) > 15 and distance_to_wall < w * 0.45:
                vehicle_risk_score = max(vehicle_risk_score, 0.95)

            if lane_change_risk:
                vehicle_risk_score = max(vehicle_risk_score, 0.6)
                if not hasattr(tr, "lane_change_warned"):
                    lane_change_warnings += 1
                    tr.lane_change_warned = True

            if tr.stationary_frames >= IDLE_FRAMES_THRESHOLD:
                vehicle_risk_score = max(vehicle_risk_score, 0.65)
                if not tr.long_idle_flagged:
                    long_idle_events += 1
                    tr.long_idle_flagged = True

            if tr.prev_speed < 2.0 and len(tr.speeds) > 1 and max(tr.speeds) > SUDDEN_STOP_SPEED_DROP:
                vehicle_risk_score = max(vehicle_risk_score, 0.7)
                if not tr.sudden_stop_flagged:
                    sudden_stop_events += 1
                    tr.sudden_stop_flagged = True

        tr.max_risk_score = max(tr.max_risk_score, vehicle_risk_score)
        display_risk = tr.max_risk_score

        if display_risk >= 0.9:
            color = (0, 0, 255)
        elif display_risk >= 0.6:
            color = (0, 215, 255)
        else:
            color = (0, 255, 0)

        cv2.rectangle(display_frame, (x1, y1), (x2, y2), color, 2)

        if tr.predicted_positions:
            for i, (px, py) in enumerate(tr.predicted_positions):
                radius = max(1, 5 - i)
                cv2.circle(display_frame, (int(px), int(py)), radius, (255, 255, 0), -1)

        if display_risk >= 0.3:
            risk_percent_text = f"{int(display_risk * 100)}%"
            label_y = max(y1 - 5, 20)
            cv2.putText(display_frame, risk_percent_text, (x1, label_y),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.7, color, 2)

    if car_count >= HIGH_DENSITY_THRESHOLD:
        high_density_events += 1

    out.write(display_frame)

cap.release()
out.release()

print("\n" + "=" * 50)
print("REPORT")
print("=" * 50)
print(f"Frames: {frame_index}")
print(f"Max Vehicles: {max_vehicles_in_frame}")
print(f"High Density Alerts: {high_density_events}")
print(f"Long Idle Alerts: {long_idle_events}")
print(f"Sudden Stop Alerts: {sudden_stop_events}")
print(f"Collision Warnings: {predicted_collisions}")
print(f"Lane Change Warnings: {lane_change_warnings}")
print("=" * 50)
print(f"Saved: {out_path}")

display(Video(out_path, embed=True, width=800))
```
